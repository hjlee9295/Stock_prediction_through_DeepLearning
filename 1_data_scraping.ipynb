#About NoteBook

This notebook is to scrape stock tick data from [dukascopy](https://www.dukascopy.com/swiss/english/marketwatch/historical/). Instead of building a scraper, we have used the [github repo](https://github.com/giuse88/duka) for scraping. The scraper has a function to scrape a period of time, but sometimes, it does not scrape the interday, resulting dataset to miss some days in the period. Our implementation is to scrape a day with a loop and save it in the google drive for later use.

```
#Stock List
#One can designate the stock name here for scraping

stock_list = ['AMZN', 'SBUX']
```

```
from datetime import datetime, timedelta
import pandas as pd
import os
import re

start_date = '2020-01-01'
day_range = 310
#stock_list = ['TSLA']

for stock in stock_list:
  stock_name = stock + 'USUSD'
  dest_name = '/content/drive/MyDrive/ADL_Final/' + stock + '_tick.csv'
  folder_name = '/content/' + stock
  for i in range(day_range):
    res = (datetime.strptime(start_date, '%Y-%m-%d') + timedelta(days=i)).strftime('%Y-%m-%d')
    !duka $stock_name -d $res -f $folder_name 
    print(res)

  file_list = os.listdir(folder_name)
  file_list.sort()
  df_dict = {'Time':[], 'Bid':[], 'Ask':[], 'Bid Volume':[], 'Ask Volume':[]}

  for file in file_list:
    read_file = folder_name + '/' + file
    if re.findall(r'\.csv', file) == []:
      continue
    try:
      df = pd.read_csv(read_file, header = None)
      df_dict['Time'] += list(df[0])
      df_dict['Bid'] += list(df[2])
      df_dict['Ask'] += list(df[1])
      df_dict['Bid Volume'] += list(df[4])
      df_dict['Ask Volume'] += list(df[3])
    except:
      pass

  final_df = pd.DataFrame.from_dict(df_dict)
  final_df.to_csv(dest_name)
```
